```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Datos
xi = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
yi = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Ajuste de polinomio de grado 1 (recta)
poly1 = PolynomialFeatures(degree=1)
X_poly1 = poly1.fit_transform(xi.reshape(-1, 1))
model1 = LinearRegression().fit(X_poly1, yi)
y_pred1 = model1.predict(X_poly1)
error1 = np.mean((yi - y_pred1) ** 2)

# Ajuste de polinomio de grado 2 (parábola)
poly2 = PolynomialFeatures(degree=2)
X_poly2 = poly2.fit_transform(xi.reshape(-1, 1))
model2 = LinearRegression().fit(X_poly2, yi)
y_pred2 = model2.predict(X_poly2)
error2 = np.mean((yi - y_pred2) ** 2)

# Ajuste de polinomio de grado 3
poly3 = PolynomialFeatures(degree=3)
X_poly3 = poly3.fit_transform(xi.reshape(-1, 1))
model3 = LinearRegression().fit(X_poly3, yi)
y_pred3 = model3.predict(X_poly3)
error3 = np.mean((yi - y_pred3) ** 2)

# Ajuste exponencial y = be^(ax)
log_yi = np.log(yi)
model_exp = LinearRegression().fit(xi.reshape(-1, 1), log_yi)
a_exp = model_exp.coef_[0]
b_exp = np.exp(model_exp.intercept_)
y_pred_exp = b_exp * np.exp(a_exp * xi)
error_exp = np.mean((yi - y_pred_exp) ** 2)

# Ajuste potencial y = bx^a
log_xi = np.log(xi)
model_pot = LinearRegression().fit(log_xi.reshape(-1, 1), log_yi)
a_pot = model_pot.coef_[0]
b_pot = np.exp(model_pot.intercept_)
y_pred_pot = b_pot * (xi ** a_pot)
error_pot = np.mean((yi - y_pred_pot) ** 2)
```

```python
# Resultados
print("Coeficientes del polinomio de grado 1:", model1.coef_, model1.intercept_)
print("Error del polinomio de grado 1:", error1)

print("Coeficientes del polinomio de grado 2:", model2.coef_, model2.intercept_)
print("Error del polinomio de grado 2:", error2)

print("Coeficientes del polinomio de grado 3:", model3.coef_, model3.intercept_)
print("Error del polinomio de grado 3:", error3)

print("Coeficientes de la función exponencial: a =", a_exp, ", b =", b_exp)
print("Error de la función exponencial:", error_exp)

print("Coeficientes de la función potencial: a =", a_pot, ", b =", b_pot)
print("Error de la función potencial:", error_pot)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Datos
xi = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
yi = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Ajuste de polinomio de grado 1 (recta)
poly1 = PolynomialFeatures(degree=1)
X_poly1 = poly1.fit_transform(xi.reshape(-1, 1))
model1 = LinearRegression().fit(X_poly1, yi)
y_pred1 = model1.predict(X_poly1)

# Ajuste de polinomio de grado 2 (parábola)
poly2 = PolynomialFeatures(degree=2)
X_poly2 = poly2.fit_transform(xi.reshape(-1, 1))
model2 = LinearRegression().fit(X_poly2, yi)
y_pred2 = model2.predict(X_poly2)

# Ajuste de polinomio de grado 3
poly3 = PolynomialFeatures(degree=3)
X_poly3 = poly3.fit_transform(xi.reshape(-1, 1))
model3 = LinearRegression().fit(X_poly3, yi)
y_pred3 = model3.predict(X_poly3)

# Ajuste exponencial y = be^(ax)
log_yi = np.log(yi)
model_exp = LinearRegression().fit(xi.reshape(-1, 1), log_yi)
a_exp = model_exp.coef_[0]
b_exp = np.exp(model_exp.intercept_)
y_pred_exp = b_exp * np.exp(a_exp * xi)

# Ajuste potencial y = bx^a
log_xi = np.log(xi)
model_pot = LinearRegression().fit(log_xi.reshape(-1, 1), log_yi)
a_pot = model_pot.coef_[0]
b_pot = np.exp(model_pot.intercept_)
y_pred_pot = b_pot * (xi ** a_pot)

# Graficar los datos y los ajustes
plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color='black', label='Datos originales')

# Graficar los ajustes
x_range = np.linspace(min(xi), max(xi), 500)
plt.plot(xi, y_pred1, label='Polinomio de grado 1')
plt.plot(xi, y_pred2, label='Polinomio de grado 2')
plt.plot(xi, y_pred3, label='Polinomio de grado 3')
plt.plot(x_range, b_exp * np.exp(a_exp * x_range), label='Exponencial')
plt.plot(x_range, b_pot * (x_range ** a_pot), label='Potencial')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Ajustes de Mínimos Cuadrados')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos
act_scores = np.array([28, 25, 28, 27, 28, 23, 28, 29, 23, 27, 29, 28, 27, 21, 28, 26, 30, 24])
average_scores = np.array([3.84, 3.21, 3.23, 3.63, 3.75, 3.53, 3.41, 3.38, 2.03, 3.75, 3.75, 1.66, 3.12, 2.96, 2.92, 3.10, 2.81])

# Reshape para sklearn
X = act_scores.reshape(-1, 1)
y = average_scores

# Modelo de regresión lineal
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

# Coeficientes de la recta
a = model.coef_[0]
b = model.intercept_

# Imprimir la ecuación de la recta
print(f"Ecuación de la recta: y = {a:.2f}x + {b:.2f}")

# Graficar los datos y la recta de ajuste
plt.figure(figsize=(10, 6))
plt.scatter(act_scores, average_scores, color='black', label='Datos originales')
plt.plot(act_scores, y_pred, color='blue', linewidth=2, label='Recta de ajuste')
plt.xlabel('Puntuación ACT')
plt.ylabel('Promedio de puntos')
plt.legend()
plt.title('Ajuste de la recta por mínimos cuadrados')
plt.show()
```
